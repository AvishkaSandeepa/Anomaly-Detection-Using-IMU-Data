```python
# import the rosbag libraries
import bagpy
from bagpy import bagreader
# import panda library for data vitualization
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# create a data folder for given b ag file
bag = bagreader("ICAB1/02_icab1_reverse_fail_overtaking0_2021-03-26-14-40-28_0.bag")
```

```python
# check the topics that are in the data folder
bag.topic_table
```

```python
# read the data of bag file using its topic
IMU_DATA = bag.message_by_topic("/icab1/mavros/imu/data")
IMU_DATA
```

```python
# displaying csv file 
IMU_DATA_vitualization = pd.read_csv(IMU_DATA)
IMU_DATA_vitualization
```

```python
# plot the details
fig , ax = plt.subplots(figsize = (20, 10))
ax.set_title("Orientation of x", fontsize=30)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Orientation of x', fontsize=18)
ax.plot(IMU_DATA_vitualization["Time"], IMU_DATA_vitualization["orientation.x"])
plt.savefig('ICAB1/02_icab1_reverse_fail_overtaking0_2021-03-26-14-40-28_0/orientaion_of_x.jpg')
#plt.plot(IMU_DATA_vitualization["Time"], IMU_DATA_vitualization["orientation.x"])
```
